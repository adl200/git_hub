Apartado A)

```python
descripcion = df.describe().T 
print(descripcion)
```

La función describe() de pandas calcula los estadísticos descriptivos básicos de cada variable numérica del conjunto de datos:

- count: número de observaciones (filas) disponibles.
- mean: media aritmética de la variable.
- std: desviación estándar, indica cuánto se dispersan los valores respecto a la media.
- min y max: valores mínimo y máximo observados.
- 25%, 50%, 75%: percentiles (cuartiles), que indican los valores por debajo de los cuales se encuentra el 25%, 50% (mediana) y 75% de los datos.

Apartado B)

```python
def clasificar_calidad(q):
    if q <= 4:
        return "malo"
    elif q <= 6:
        return "regular"
    else:
        return "bueno"

df["quality_cat"] = df["quality"].apply(clasificar_calidad)
```
