La función describe() de pandas calcula los estadísticos descriptivos básicos de cada variable numérica del conjunto de datos:

- count: número de observaciones (filas) disponibles.
- mean: media aritmética de la variable.
- std: desviación estándar, indica cuánto se dispersan los valores respecto a la media.
- min y max: valores mínimo y máximo observados.
- 25%, 50%, 75%: percentiles (cuartiles), que indican los valores por debajo de los cuales se encuentra el 25%, 50% (mediana) y 75% de los datos.

Ejercicio Nº4

```python
estudiantes = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "nombre": ["Ana", "Luis", "Marta", "Jorge"]
})

notas = pd.DataFrame({
    "id": [1, 2, 2, 5],
    "asignatura": ["Matemáticas", "Historia", "Química", "Biología"],
    "nota": [8.5, 7.0, 6.5, 9.0]
})
```

```python
print(estudiantes)
```

```python
print(notas)
```

Apartado A)

```python
merge_inner = pd.merge(estudiantes, notas, on="id", how="inner")
print(merge_inner)
print("Número de filas:", len(merge_inner))
```

En el merge inner se obtienen 3 filas porque solo se combinan los ids comunes (1 y 2).
El id 1 genera una fila y el id 2 genera dos filas al tener dos notas distintas en el DataFrame notas.

Apartado B)

```python
merge_outer = pd.merge(estudiantes, notas, on="id", how="outer")
print(merge_outer)
```

En el outer se incluyen todos los ids que aparecen en cualquiera de los dos DataFrames:
- id 1 y 2: están en ambos, salen con nombre, asignatura y nota.
- id 3 (Marta) y id 4 (Jorge): solo existen en estudiantes, aparecen con NaN en asignatura y nota porque no tienen notas asociadas.
- id 5 (Biología, 9.0): solo existe en notas, aparece con NaN en nombre porque no hay estudiante con ese id.

Apartado C)

```python
merge_left = pd.merge(estudiantes, notas, on="id", how="left")
print(merge_left)
```

En el left merge se mantienen todos los estudiantes (ids 1–4), por lo que Marta y Jorge aparecen aunque no tengan notas (con NaN en las columnas de notas).
La única información que se pierde respecto al outer es la fila con id 5 (Biología, 9.0), que solo estaba en notas y no tiene estudiante asociado.

Apartado D)

```python
merge_right = pd.merge(estudiantes, notas, on="id", how="right")
print(merge_right)
```

En el right merge usando notas como tabla derecha se mantienen todas las filas de notas (ids 1, 2, 2 y 5).
Los estudiantes con id 3 (Marta) e id 4 (Jorge) desaparecen del resultado porque no tienen ninguna nota asociada.
Además, la fila con id 5 aparece con NaN en nombre ya que no existe ningún estudiante con ese id.

Apartado E)

```python
estudiantes["curso"] = ["1º ESO", "2º ESO", "3º ESO", "4º ESO"]
print(estudiantes)
```

```python
merge_curso = pd.merge(estudiantes, notas, on="id", how="inner")
print(merge_curso)
```

He añadido la columna curso al DataFrame estudiantes y después he hecho un merge interno por id.
En el resultado aparecen solo los estudiantes que tienen notas (id 1 y 2).

- Ana (1º ESO) aparece una vez con su asignatura y nota.
- Luis (2º ESO) aparece dos veces, una por cada asignatura (Historia y Química).
Esto muestra un caso de relación uno-a-muchos: un estudiante puede tener varias filas en el resultado porque está matriculado en varias asignaturas.
